```python
def crypte(data, columns, mdp):
    
    for i in columns:
        crypte = []
        
        for u in data[i]:
            crypte.append(cryptocode.encrypt(str(u), "mdp"))
            
        data[i] = crypte
    
    return data
```

# sql operations

```python
engine = create_engine(database_url)
```

```python
with engine.connect() as conn:

    conn.execute(CreateSchema("salesdatamart", if_not_exists=True))
    conn.execute(CreateSchema("inventorydatamart", if_not_exists=True))
    conn.commit()
```

```python
dateh.to_sql('date', schema="salesdatamart", con=engine, if_exists="replace", index=False)
dateh.to_sql('date', schema="inventorydatamart", con=engine, if_exists="replace", index=False)
```

```python
customers.to_sql('customers', schema="salesdatamart", con=engine, if_exists="replace", index=False)
```

```python
products.to_sql('products', schema="inventorydatamart", con=engine, if_exists="replace", index=False)
products.to_sql('products', schema="salesdatamart", con=engine, if_exists="replace", index=False)
```

```python
suppliers.to_sql('suppliers', schema="inventorydatamart", con=engine, if_exists="replace", index=False)
```

```python
shippers.to_sql('shippers', schema="salesdatamart", con=engine, if_exists="replace", index=False)
```

```python
transactions.to_sql('transactions', schema="salesdatamart", con=engine, if_exists="replace", index=False)
```

```python
stock.to_sql('stock', schema="inventorydatamart", con=engine, if_exists="replace", index=False)
```

```python
salesfact.to_sql('salesfact', schema="salesdatamart", con=engine, if_exists="replace", index=False)
```

```python
inventoryfact.to_sql('inventoryfact', schema="inventorydatamart", con=engine, if_exists="replace", index=False)
```
