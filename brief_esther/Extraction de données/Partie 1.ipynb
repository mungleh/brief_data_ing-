```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
```

```python
def get_soup(url):
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    return soup
```

```python
def get_rows(table, crit_1=None, crit_2=None, crit_3=None, extract_type=None, atrs_1=None, atrs_2=None):
    
    truc = []
    
    if extract_type == 1:
        for u in table.find_all(crit_1):
            for i in u.find_all(crit_2):
                row_data = []
                for cell in i.find_all(crit_3):
                    row_data.append(cell.text)
                truc.append(row_data)
                
    elif extract_type == 2:
        for i in table.find_all(crit_1):
            row_data = []
            for cell in i.find_all(crit_2):
                row_data.append(cell.text)
            truc.append(row_data)
            
    elif extract_type == 3:
        for i in table.find_all(crit_1, atrs_1):
            row_data = []
            for cell in i.find_all(crit_2):
                row_data.append(cell.text)
            truc.append(row_data)
            
    elif extract_type == 4:
        for i in table.find_all(crit_1, atrs_1):
            row_data = []
            for tag in i.find_all(crit_2, atrs_2):
                row_data.append(tag['data-rating'])
            for cell in i.find_all(crit_3):
                row_data.append(cell.text)
            truc.append(row_data)
    
    return truc
```

```python
def lol(url):
    
    # get page numbers
    table = get_soup(url)

    data = get_rows(table,
                    crit_1="div",
                    crit_2="li",
                    extract_type=3,
                    atrs_1={"id": "static-pagination"})

    df = pd.DataFrame(data)
    df = df.T
    maxa = df[0].str.extract('(\d+)', expand=False).explode().astype(float).max()
    
    # get pages data
    doto = []

    for i in range(1, int(maxa)+1):
        putin = get_soup(f'{url}?page={i}')
        
        diti = get_rows(putin,
                        crit_1="div",
                        crit_2='p',
                        crit_3=['a', 'h4', 'p'],
                        extract_type=4,
                        atrs_1={"class": "product-wrapper card-body"},
                        atrs_2={"data-rating" :  [1,2,3,4,5]})
        
        for i in diti:
            doto.append(i)

    dfe = pd.DataFrame(doto)
    dfe = dfe.apply(lambda col: col.str.replace('$', ''))
    dfe = dfe.drop([dfe.columns[2], dfe.columns[5], dfe.columns[6]], axis=1)
    dfe = dfe.rename({0: "ratings", 1: "prix", 3: "name", 4: "description"}, axis= 1)
    dfe[['ratings','prix']] = dfe[['ratings','prix']].apply(pd.to_numeric)

    return dfe
```

```python
pc = lol("https://webscraper.io/test-sites/e-commerce/static/computers/laptops")
tablait = lol("https://webscraper.io/test-sites/e-commerce/static/computers/tablets")
```
